```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Initialise data to lists
total = len(purchase_data.index)
data = [
    {
        'Total Players': total
    }
] 

# Create DataFrame
df = pd.DataFrame(data) 
  
# Print the data 
df 
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.describe()
items = purchase_data["Item ID"].unique()

num_unique_items = len(items)
avg_price = purchase_data["Price"].mean()
num_purchases = len(purchase_data["Item ID"])
total_revenue = purchase_data["Price"].sum()

# Initialise data to lists
data = [
    {
        'Number of Unique Items': num_unique_items, 
        'Average Price': avg_price, 
        'Number of Purchases': num_purchases, 
        'Total Revenue': total_revenue
    }
] 
  
# Create DataFrame
df = pd.DataFrame(data) 

# Format
df["Average Price"] = df["Average Price"].map("${0:,.2f}".format)
df["Total Revenue"] = df["Total Revenue"].map("${0:,.2f}".format)
  
# Print the data 
df 
```

## Top Spenders

* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_df = purchase_data.groupby(["SN"])

# Calculate total purchases for each SN
spenders_purchase = spenders_df["Price"].sum()

# Calculate purchase count for each spender
purchase_count = spenders_df["Purchase ID"].count()

# Calculate average purchase for each spender
avg_purchase_price = spenders_df["Price"].mean()

# Create summary DataFrame
spenders_data = pd.DataFrame(
    {
        "Purchase Count": purchase_count,
        "Average Purchase Price": avg_purchase_price,
        "Total Purchase Value": spenders_purchase
    }
)

# Sort
top_spenders = spenders_data.sort_values(["Total Purchase Value"], ascending = False)

# Format
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${0:,.2f}".format) 
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${0:,.2f}".format) 

top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data.groupby(["Item ID","Item Name"])

# Calculate purchase count for each purchase ID
purchase_count = popular_items["Purchase ID"].count()

# Calculate total price for each group
total_value = popular_items["Price"].sum()

item_price = total_value / purchase_count

# Create summary DataFrame
top_items = pd.DataFrame(
    {
        "Purchase Count": purchase_count,
        "Item Price": item_price,
        "Total Purchase Value": total_value
    }
)

# Sort
popular_df = top_items.sort_values(["Purchase Count"], ascending = False)

# Format
popular_df["Item Price"] = popular_df["Item Price"].map("${0:,.2f}".format)
popular_df["Total Purchase Value"] = popular_df["Total Purchase Value"].map("${0:,.2f}".format)
                                                          
popular_df.head()
```

## Most Profitable Items
